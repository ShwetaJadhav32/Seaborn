```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
df
```

```python
sns.stripplot(data=df,x=df['Species'],y=df['Sepal_Width'])
```

```python
# sns.stripplot([x,y])  ## Here we can not use stripplot because it takes numerical values.
```

```python
# sns.stripplot(df['Species'],df['Sepal_Width'])
```

```python
sns.stripplot(data=df,x=df['Species'],y=df['Sepal_Length'],palette=['g','b','r'])
plt.title("Strip plot")
plt.xlabel("Species")
plt.ylabel("Sepal_Width")
plt.show()
```

```python
sns.barplot(data=df,x=df['Species'],y=df['Sepal_Length'],palette='Blues_d')
plt.title("Bar plot")
plt.xlabel("Species")
plt.ylabel("Sepal_Width")
plt.show()
```

```python
sns.countplot(data=df,x=df['Sepal_Length'])
```

```python
sns.boxplot(data=df,x=df['Species'],y=df['Sepal_Width'])
```

```python
sns.boxenplot(data=df,x=df['Species'],y=df['Sepal_Width'])
```

```python
sns.violinplot(data=df,x=df['Species'],y=df['Sepal_Width'])
```
