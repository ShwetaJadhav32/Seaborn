#### Refer to the tips.csv dataset. The dataset contains 244 tips(data about customers served by a restaurant)across 7 different columns.

* **totbill** - Total bill (cost of the meal), including tax, in US dollars
* **tip** - Tip (gratuity) in US dollars
* **sex** - Sex of person paying for the meal (0=male, 1=female)
* **smoker** - Smoker in party? (0=No, 1=Yes)
* **day** - Sun, Sat, Thu, Fri
* **time** - Night and Day
* **size** - Size of the party

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('tips.csv')
df
```

```python
df.info()
```

### Q1.Plot a day wise boxplot of totbill column as shown below.
Note - In all the visualizations here, the rainbow palette is used.

```python
# sns.boxplot(x="day",y="total_bill",data=df,palette="rainbow")   OR
sns.boxplot(data=df,x=df['day'],y=df['total_bill'],palette='rainbow')
plt.title("Day wise Boxplot of totbill")
plt.show()
```

### Q2. Plot a pairplot by setting hue parameter to sex

```python
sns.pairplot(data=df,hue="sex",palette="rainbow")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="tip",y="total_bill",hue="sex",data=df,palette="Set1")
plt.title("Scatterplot of tip vs. total_bill")
plt.show()
```

### Q4. Plot a histogram of totbill for 50 bins.

```python
sns.distplot(df["total_bill"],bins=50,kde=True)
plt.title("Histogram of total_bill")
plt.show()
```

### Q5. Plot a heatmap to represent entire dataset's correlation.

```python
df1=df.select_dtypes(include=['float64', 'int64'])
df1
```

```python
sns.heatmap(df1.corr(),cmap="rainbow",annot=True)
plt.title("tips data correlation")
plt.show()
```
